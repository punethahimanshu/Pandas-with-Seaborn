```python
## Data to plot in pie chart ##

labels = 'GATE','GRE','UPSC','UKPCS'
sizes = [91,340,100,200]
colors = ['red', 'green','lightblue','yellow']
explode = (0.1,0,0,0) ## explode slices the pie chart

#plot
plt.pie(sizes, labels=labels,colors=colors,explode=explode,autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
```

# corelation

corelation with heatmap

```python
## if you want to find the corelation between different features then use <<df.corr()>> and then heat map ###
### you can only find corelation between integers and floats not with category ### REMEMBER THIS POINT


df.corr()
```

```python
sns.heatmap(df.corr())
```

# Bivariate Analysis

Two Variables are involved

```python
sns.jointplot(x ='tip', y='total_bill', data=df,kind='reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
df['smoker'].value_counts()    ### to find how many are smokers ### df1[<name>].value_counts() ###
```

```python
sns.pairplot(df,hue='smoker')
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde = False, bins =10)
```

# Catogrical Plot

## **seaborn also helps in doing analysis of Catogrical Data points:-**

* boxplot
* violinplot
* countplot
* barplot

```python
### countplot ###
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df['sex'].value_counts()
```

```python
sns.countplot('sex',data = df)
```

```python
df['day'].value_counts()
```

```python
sns.countplot(y='day',data=df)
```

```python
### barplot ###
```

```python
sns.barplot(y ='total_bill',x='sex', data = df)
```

```python
### Boxplot ###
```

```python
sns.boxplot(y='total_bill',x='sex',data=df)
```

```python
sns.boxplot(x='day',y='total_bill', data=df,palette='rainbow')
```
